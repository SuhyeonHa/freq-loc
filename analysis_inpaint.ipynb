```python
import torch
import matplotlib.pyplot as plt

data = torch.load("inpaint_history.pt")
mse_inside = data['mse_inside_history']
mse_outside = data['mse_outside_history']
timesteps = data['timesteps']

plt.figure(figsize=(10, 5))
plt.plot(timesteps.numpy(), mse_inside, label="MSE Inside Mask (Creation)", marker='o')
plt.plot(timesteps.numpy(), mse_outside, label="MSE Outside Mask (Restoration)", marker='x')
plt.xlabel("Timestep")
plt.ylabel("Mean Squared Error (MSE)")
plt.title("MSE between Predicted Noise and Actual Noise")
plt.legend()
plt.grid(True)
# plt.yscale('log')
plt.gca().invert_xaxis()
plt.savefig("mse_inpaint_latent.png", bbox_inches='tight')
plt.show()
```
